## Scaling and Train Test Split

```python
# Now begins the training of model 
X = df.drop('price',axis=1).values
y = df['price'].values
# .values will return the numpy array
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train) # Same transform 
```

```python
X_test = scaler.transform(X_test) 
# We dont fit our test state we only transform it 
```

```python
X_train
```

```python
X_test
```

```python
X_train.shape
```

```python
X_test.shape
```

## Creating a Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
# WE have 19 features so we will be using 19 nodes
# We are making a deep neural network of 4 hidden units

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

# THis is the outcome which will predict the price
model.add(Dense(1))

# Now we are using ADAM Optimizer and ofc our loss will be mse as it a regression problem
model.compile(optimizer='adam',loss='mse')
```

## Training the Model

```python
# Use validation_data it will just check that we are working fine with our model 
# batch_size try to keep it in powers of 2 and small the batch larger the size 
# If we keep small batches we might overfit the data 
# epochs will just run it multiple times 
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128,epochs=400)

```

```python
losses = pd.DataFrame(model.history.history)
losses
```

```python
losses.plot()
# If the graph goes above then we are Overfitting the data but the graph is smooth 
# i.e it is a good song 
```
